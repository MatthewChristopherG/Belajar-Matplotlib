# Simple Pie Plot

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
```

```python
plt.pie(data,
       labels=kategori,
       autopct='%1.lf%%',
       startangle=90)

plt.title('Contoh Pie Plot')

plt.show()
```

# Pengaturan Legend can Color

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue']
```

```python
plt.pie(data,
      labels=kategori,
        colors=warna,
       autopct='%1.lf%%',
       startangle=90)

plt.legend()
plt.title('Contoh Pie Plot')

plt.show()
```

```python
plt.pie(data,
      labels=kategori,
        colors=warna,
       autopct='%1.lf%%',
       startangle=80)

plt.legend()
plt.title('Contoh Pie Plot')

plt.show()
```

Start angle makin kecil berputar sesuai arah jarumjam, bila semakin besar melawan arah jarum jam

# Exploded Pie Plot

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue']
explode_var =  [0., 0., 0.2, 0., 0.08]
```

```python
plt.pie(data,
      labels=kategori,
        colors=warna,
        explode=explode_var,
       autopct='%1.lf%%',
       startangle=90)

plt.legend()
plt.title('Contoh Exploded Pie Plot')

plt.show()
```

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue']
explode_var =  [0., 0., 0.1, 0., 0.08]
```

```python
plt.pie(data,
      labels=kategori,
        colors=warna,
        explode=explode_var,
       autopct='%1.lf%%',
       startangle=90)

plt.legend()
plt.title('Contoh Exploded Pie Plot')

plt.show()
```
