# Simple Bar Plot

```python
data = [25, 45, 55, 125, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori, data)
plt.grid()

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

# Pengaturan Grid dan Color

```python
data = [25, 45, 55, 125, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori,
       data,
       color='red',
       alpha=0.25) # alpha merupakan untuk mengatur tingkat transparansi

plt.grid(linestyle='--',
        axis='y', # pilihan: x, y, both
        alpha=0.50)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

```python
Stacked Bar Plot
```

```python
data1 = [25, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori, data1, label='Data 1')
# plt.bar(kategori, data2, label='Data 2')
plt.bar(kategori, data2, label='Data 2', bottom=data1)

plt.legend()

plt.grid(linestyle='-',
        linewidth=1,
        alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Stacked Bar Plot')

plt.show()
```

```python
plt.bar(kategori, data1, label='Data 1', bottom=data2)
# plt.bar(kategori, data2, label='Data 2')
plt.bar(kategori, data2, label='Data 2')

plt.legend()

plt.grid(linestyle='-',
        linewidth=1,
        alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Stacked Bar Plot')

plt.show()
```

# Grouped Bar Plot

```python
data1 = [25, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ['A', 'B', 'C', 'D', 'E']

x = np.arange(len(kategori))
x
```

```python
width = 0.35

plt.bar(x-width/2, data1, width, label='Data1')
plt.bar(x+width/2, data2, width, label='Data2')

plt.xticks(x, kategori)

plt.grid(linestyle='-',
        linewidth=1,
        axis='y',
        alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Stacked Bar Plot')

plt.legend()
plt.show()
```

```python
width = 0.5

plt.bar(x-width/2, data1, width, label='Data1')
plt.bar(x+width/2, data2, width, label='Data2')

plt.xticks(x, kategori)

plt.grid(linestyle='-',
        linewidth=1,
        axis='y',
        alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Stacked Bar Plot')

plt.legend()
plt.show()
```
